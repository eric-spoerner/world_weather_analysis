## Make a heatmap of locations

### Start with temperature

```python
# Assign locations to an array of lat/long pairs
locations = city_data_df[["Lat", "Long"]]

# Assign weights variable to some values
max_temp = city_data_df["Max Temp"]
        
#Assign a fig to the gmaps.figure() attribute
fig = gmaps.figure(center=(0.0, 0.0), 
                   zoom_level=1.5,
                   layout = {'width': '800px',
                             'height': '500px'})

# #assign heatmap layer variable t o heatmap layer attrib and add other dfs
heat_layer = gmaps.heatmap_layer(locations,
                                 weights=[max(temp,0) for temp in max_temp],
                                 dissipating=False,
                                 point_radius=3,
                                 max_intensity=0.95)

# # #Add the heatmap layer.
fig.add_layer(heat_layer)

# #Call the figure
fig
```

### Humidity
